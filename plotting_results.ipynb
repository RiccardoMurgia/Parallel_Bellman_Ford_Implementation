# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading data

```python
strong_efficiency_data = pd.read_csv("output1.txt", sep=" ")
weak_efficiency_data = pd.read_csv("output2.txt", sep=" ")
openMp_VS_cuda_data = pd.read_csv("openMp_VS_cuda.txt", sep=" ")
cuda_without_copy = pd.read_csv("cuda_without_copy.txt", sep=" ")

# You need to delete the last column because when reading txt files, a column of NaN is added because of the separator " "
#strong_efficiency_data = strong_efficiency_data.drop(strong_efficiency_data.columns[-1], axis=1)
#weak_efficiency_data = weak_efficiency_data.drop(weak_efficiency_data.columns[-1], axis=1)
openMp_VS_cuda_data = openMp_VS_cuda_data.drop(openMp_VS_cuda_data.columns[-1], axis=1)
cuda_without_copy = cuda_without_copy.drop(cuda_without_copy.columns[-1], axis=1)

strong_efficiency_data = strong_efficiency_data.set_index('Algorithm')
weak_efficiency_data = weak_efficiency_data.set_index('Algorithm')
openMp_VS_cuda_data = openMp_VS_cuda_data.set_index('Algorithm')
cuda_without_copy['Algorithm'] = cuda_without_copy['Algorithm'].apply(lambda x:x+"_without_copy")
cuda_without_copy = cuda_without_copy.set_index('Algorithm')


display(strong_efficiency_data)
#display(weak_efficiency_data)
openMp_VS_cuda_data = pd.concat([openMp_VS_cuda_data, cuda_without_copy], axis=0)
display(openMp_VS_cuda_data)
```

# Plotting OpenMp Times slurm

```python
strong_efficiency_data_mean = strong_efficiency_data.iloc[:, 0::2]
strong_efficiency_data_std_dev = strong_efficiency_data.iloc[:, 1::2]

plt.figure(figsize=(12, 6))

colors = ['red', 'darkblue', 'pink', 'orange', 'purple', 'blue', 'green']
for i in range(strong_efficiency_data.shape[0]):
    algorithm = str(strong_efficiency_data_mean.index[i])
    mean_values = strong_efficiency_data_mean.iloc[i, :]
    std_dev_values = strong_efficiency_data_std_dev.iloc[i, :]
    
    plt.plot(range(1, strong_efficiency_data_mean.shape[1]+1), strong_efficiency_data_mean.iloc[i, :], label=algorithm + " Mean", color=colors[i])
    plt.fill_between(range(1, strong_efficiency_data_mean.shape[1]+1), 
                     mean_values.values -  std_dev_values.values, 
                     mean_values.values +  std_dev_values.values, 
                     color=colors[i], 
                     alpha=0.2, 
                     label=algorithm + " Std Dev")
 
plt.title("Execution Times of a Graph with V=1000 as N Threads Increase")
plt.xlabel("Number of Threads")
plt.ylabel("Wall clock time (seconds)")
plt.legend()    
plt.grid()
```

# Strong Efficiency and Speed Up


```python
times_using_one_core = strong_efficiency_data_mean['mean_TH_1']

speedup = strong_efficiency_data_mean.copy()
strong_efficiency = strong_efficiency_data_mean.copy()
p = 2

for i in strong_efficiency_data_mean.columns:
    speedup.loc[:, i] = times_using_one_core / strong_efficiency_data_mean.loc[:, i]
    strong_efficiency.loc[:, i] = speedup.loc[:, i] / p
    p += 1

colors = ['red', 'darkblue', 'pink', 'orange', 'purple', 'blue', 'green']

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

for i in range(1, speedup.shape[0]):
    algorithm = str(speedup.index[i])
    axs[0].plot(range(1, speedup.shape[1] + 1), speedup.iloc[i, :], label=algorithm, color=colors[i])
axs[0].set_title("Speed Up varying the number of thread")
axs[0].set_xlabel("Number of Threads", fontsize=12)
axs[0].set_ylabel("Speed Up",  fontsize=12)
axs[0].legend(loc='lower right')
axs[0].grid()

for i in range(1, strong_efficiency.shape[0]):
    algorithm = str(strong_efficiency.index[i])
    axs[1].plot(range(1, strong_efficiency.shape[1] + 1), strong_efficiency.iloc[i, :], label=algorithm, color=colors[i])
axs[1].set_title("Strong Efficiency")
axs[1].set_xlabel("Number of Threads",  fontsize=12)
axs[1].set_ylabel("Strong Efficiency", fontsize=12)
axs[1].legend(loc='lower right')
axs[1].grid()

# Mostra il grafico
plt.tight_layout()
plt.show()
```

## OpenMP vs CUDA

```python
sizes = [2000, 5000, 10000]
openMp_VS_cuda_data_new = openMp_VS_cuda_data.iloc[:11 ,6:].copy()
openMp_VS_cuda_data_mean = openMp_VS_cuda_data_new.iloc[:, 0::2]
throughput = openMp_VS_cuda_data_mean.copy()
for index, col_name in enumerate(openMp_VS_cuda_data_mean.columns):
    if col_name == "V0" or col_name == "V0_1" or col_name == "cuda_V0" or col_name == "cuda_V0_1":
        throughput.loc[:, col_name] = (sizes[index]**3 * 3 * 2) / openMp_VS_cuda_data_mean.loc[:, col_name]
    else:
        throughput.loc[:, col_name] = (sizes[index]**3 * 2) / openMp_VS_cuda_data_mean.loc[:, col_name]
        
        
colors = plt.cm.tab20(np.linspace(0, 1, 11))
plt.figure(figsize=(12, 6))

for i in range(throughput.shape[0]):
    algorithm = str(throughput.index[i])
    plt.plot(sizes, throughput.iloc[i, :], label=algorithm, color=colors[i])

plt.title("Throughput")
plt.xlabel("Number of Nodes")
plt.ylabel("Element processed per second (B/S)")

plt.legend()    
plt.grid()
```

## Analysis of the percentage of time needed to copy the graph to gpu versus the total execution time 

```python
openMp_VS_cuda_data_mean = openMp_VS_cuda_data.iloc[:, 0::2]

mask = (openMp_VS_cuda_data_mean.index == 'cuda_V1_1') | \
       (openMp_VS_cuda_data_mean.index == 'cuda_V1_1_without_copy')

tmp = openMp_VS_cuda_data_mean[mask]
percentage_of_copy = 1 - (tmp[tmp.index == 'cuda_V1_1_without_copy'].values / tmp[tmp.index == 'cuda_V1_1'].values)  # senza copia/ copia

plt.figure(figsize=(12, 6))
plt.plot([100, 500, 1000, 2000, 5000, 10000], percentage_of_copy.flatten())
plt.title("Time to execute the kernels / Total time")
plt.xlabel("Number of Nodes")
plt.ylabel("% Wall clock time (seconds)")
plt.grid()
plt.show()
```
